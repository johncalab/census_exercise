```python
# packages
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks", context="talk")
plt.style.use("dark_background")
```

```python
from utils import load_data_from_two
df = load_data_from_two('census-income.data','census-income.columns')
```

```python
fig_size_x = 24
fig_size_y = 12

plt.figure(figsize=(fig_size_x,fig_size_y))
plt.yscale('log')
sns.countplot(df['hispanic origin'])
plt.title('hispanic origin')
sns.despine()
plt.savefig('path.png')
```

```python
len(df[df['label'] == '50000+.'])/len(df)
```

```python
# use logarithmic scale

fig_x=12
fig_y=12

plt.figure(figsize=(fig_x,fig_y))
class_0 = df[df['label'] == "- 50000."]
class_1 = df[df['label'] == "50000+."]

plt.hist(class_0['age'],bins=22)
plt.hist(class_1['age'],bins=22)
plt.xlabel('age')
plt.yscale('log')
plt.xlim(0)
sns.despine()

plt.legend(['less than 50k', 'more than 50k'])
```

```python
fig_x=12
fig_y=12

feat = 'dividends from stocks'

n_bins = 40

plt.figure(figsize=(fig_x,fig_y))
class_0 = df[df['label'] == "- 50000."]
class_1 = df[df['label'] == "50000+."]




plt.hist(class_1[feat],bins=n_bins,color='orange')
plt.hist(class_0[feat],bins=n_bins,color='blue')

plt.xlabel(feat)

plt.yscale('log')
plt.xlim(0)
sns.despine()

plt.legend(['less than 50k', 'more than 50k'])
```

```python
fig_x=12
fig_y=12

feat = 'year'

n_bins = 30

plt.figure(figsize=(fig_x,fig_y))
class_0 = df[df['label'] == "- 50000."]
class_1 = df[df['label'] == "50000+."]

plt.hist(class_0[feat],bins=n_bins)
plt.hist(class_1[feat],bins=n_bins)

plt.xlabel(feat)

plt.yscale('log')
# plt.xlim(0)
sns.despine()


plt.legend(['less than 50k', 'more than 50k'])
```

```python
fig_x = 12
fig_y = 12
col_name = 'enroll in edu inst last wk'

fig,ax = plt.subplots(1,1,figsize=(fig_x,fig_y))

df_plot = df.groupby(['label', col_name]).size().reset_index().pivot(columns='label', index=col_name, values=0)
df_plot.plot(kind='barh', stacked=True, ax=ax)

plt.legend(['less than 50k', 'more than 50k'])

sns.despine()
```
